```python
from sklearn.model_selection import KFold
import json
import os
from datasets import concatenate_datasets

def cross_validation_train(peft_model, tokenizer, lora_args, train_data, eval_data, n_splits=5):
    """
    使用交叉验证进行训练。
    :param peft_model: PEFT 模型
    :param tokenizer: 分词器
    :param lora_args: 训练参数
    :param train_data: 训练数据
    :param eval_data: 验证数据
    :param n_splits: 交叉验证的折数
    """
    # 合并训练集和验证集
    full_data = concatenate_datasets([train_data, eval_data])
    results = []
    # 使用 KFold 进行交叉验证
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)
    
    for fold, (train_idx, val_idx) in enumerate(kf.split(full_data)):
        print(f"Training fold {fold + 1}/{n_splits}...")
        
        
        # 划分训练集和验证集
        train_fold = full_data.select(train_idx)
        val_fold = full_data.select(val_idx)
        
        # 为当前 fold 生成唯一的 output_dir
        fold_output_dir = os.path.join(lora_args.output_dir, f"fold_{fold}")
        os.makedirs(fold_output_dir, exist_ok=True)
        # 复制训练参数并更新 output_dir
        fold_lora_args = TrainingArguments(
            **{**lora_args.to_dict(), "output_dir": fold_output_dir}
        )
        """
        **lora_args.to_dict()：解包原始参数字典。
        "output_dir": fold_output_dir：添加或覆盖 output_dir 的值。
        ** new_args_dict 将字典解包为关键字参数，传递给 TrainingArguments 的构造函数。
        """
        # 构建 Trainer
        trainer = build_trainer(peft_model, tokenizer, lora_args, train_fold, val_fold)

        # 开始训练
        train_result = trainer.train()
        
        print(f"fold={fold}, result = {train_result}")
        results.append(train_result)
```

# 优化策略篇

## 1. 数据增强
目的是提高模型的泛化能力，进行数据扩充。同时通过增加新的数据，可以有效防止过拟合。
### 同义词替换：替换非实体词（如“治疗” → “医治”），保留实体不变。

```python
#### 替换非实体词汇：
"""
保持句子语义不变的同时增加数据多样性，防止模型过拟合到特定的非实体词汇。
通过引入同义词，间接增加训练数据的多样性，尤其适用于小规模数据集。
"""
synonyms = {"治疗": "医治", "改善": "缓解"}  
def replace_synonyms(text):  
    return [synonyms.get(word, word) for word in text]  
```

```python
#### 随机遮盖（Masking）：随机遮盖非实体词（如“的”、“是”），强制模型关注实体上下文。
#对于实体识别：我们更关注的模型针对特殊名词的 “识别” 能力。所以对上下文的关注可能不是那么重要。
"""
随机遮盖非实体词（如“的”、“是”），强制模型依赖上下文推断实体，减少对高频但无信息量词汇的关注。
模拟真实场景：实际文本中可能存在拼写错误或省略，遮盖训练使模型更适应不完整输入。
"""
def random_mask(text, mask_rate=0.1):
    return [word if np.random.rand() > mask_rate else "[MASK]" for word in text]
```

```python
# 实体替换：在同类实体中替换（如“当归” → “黄芪”），需确保替换后的实体类型一致。
# 
"""
防止实体过拟合：模型不再依赖特定实体词（如“当归”），而是学习实体类型的通用特征（如中药的上下文模式）。若
某些实体（如“当归”）出现频率过高，替换可缓解数据不平衡问题。

我举一个例子：模型能识别未见过的同类实体（如“川芎”），提升在真实场景中的适用性。

"""
entity_mapping = {"当归": "黄芪", "丹参": "川芎"}
def replace_entity(text):
    return [entity_mapping.get(word, word) for word in text]
```

## 2.标签平滑（Label Smoothing）

调整标签的分布，减少模型对标签的过度自信，从而提高模型的泛化能力

在分类任务中，通常使用 One-Hot Encoding表示标签。例如三分类任务，真实标签为类别1，标签向量为 [1, 0, 0]。

问题：
过度自信：模型会倾向于将预测概率集中在真实类别上（如 [0.99, 0.01, 0]），而忽略其他类别的可能性。  
过拟合风险：模型可能过于依赖训练数据中的特定模式，导致在验证集或测试集上表现不佳。

```python
#代码原理：
import torch
import torch.nn as nn

class LabelSmoothingLoss(nn.Module):
    def __init__(self, smoothing=0.1):
        super(LabelSmoothingLoss, self).__init__()
        self.smoothing = smoothing
        self.confidence = 1.0 - smoothing

    def forward(self, logits, labels):
        num_classes = logits.size(-1)
        # 将标签转换为独热编码
        one_hot = torch.zeros_like(logits).scatter(1, labels.unsqueeze(1), 1)
        # 应用标签平滑
        smooth_labels = one_hot * self.confidence + (1 - one_hot) * self.smoothing / (num_classes - 1)
        # 计算交叉熵损失
        log_probs = torch.nn.functional.log_softmax(logits, dim=-1)
        loss = - (smooth_labels * log_probs).sum(dim=-1).mean()
        return loss

# 示例用法
logits = torch.tensor([[1.0, 2.0, 3.0], [1.0, 2.0, 3.0]], requires_grad=True)
labels = torch.tensor([0, 1])
loss_fn = LabelSmoothingLoss(smoothing=0.1)
loss = loss_fn(logits, labels)
print(loss)
```

## 3.动态批处理（Dynamic Batching）

注意：我们之前的步骤都是使用  固定的batch 进行训练的！
当输入数据的长度不一致时（如文本序列），需要对较短序列进行填充（Padding）以对齐批次中的最长序列。

存在弊端：在直觉上，很短的也进行填充是否有点浪费了？   （注意在预训练模型的时候，文本都被切割为 固定长度的片段了，可能不存在这个问题，但是我们现在是在做 命名体识别，手里的数据就是长度不一的所以存在这个问题）  
填充部分不参与实际计算，但会占用内存和计算资源。

优化思想：
​按长度排序：将训练数据按序列长度排序。  
​动态分组：根据序列长度将样本分组，确保每组内的序列长度相近。  
​动态填充：对每组内的序列进行填充，填充量仅基于组内的最长序列。  
​批次生成：将分组后的样本组合成批次，每个批次的填充量最小化。  



根据序列长度调整批次大小，提升训练效率。
适用于处理变长序列数据（如文本、音频等），因为传统的固定批次大小可能导致大量填充（Padding），浪费计算资源。
